**Basic usage of Shapely library**

```python
from shapely.geometry import Point, LineString
```

```python
# Create two points
point_a = Point(0,0)
point_b = Point(1,1)

# point_b wkt
print("point_b wkt: {}".format(point_b.wkt))
```

```python
# point_b X,Y coordinates
print("point_b X coordinate: {}".format(point_b.x))
print("point_b Y coordinate: {}".format(point_b.y))
```

```python
# point_b coordinate values
point_b_coords = point_b.coords[:]
print("point_b coordinate values: {}".format(point_b_coords))
```

```python
# Distance point_a - point_b
print("Distance point_a - point_b: {}".format(point_a.distance(point_b)))
```

```python
# Create a line from point_a - point_b
line_ab = LineString([point_a, point_b])
print("line_ab wkt: {}".format(line_ab.wkt))
line_ab
```

```python
# line_ab coordinate values
line_ab_coords = line_ab.coords[:]
print("line_ab coordinate values: {}".format(line_ab_coords))
```

```python
# line_ab start point
st_pt = line_ab_coords[0]
end_pt = line_ab_coords[-1]
print("Start point {} end point: {}".format(st_pt, end_pt))
```

```python
# line_ab Length
print("line_ab Length: {}".format(line_ab.length))
```

```python
# line_ab envelope
line_ab_env = line_ab.envelope
print("line_ab envelope: {}".format(line_ab_env))
line_ab_env
```

```python
# point_a buffer
point_a_buff = point_a.buffer(2)
# point_b buffer
point_b_buff = point_b.buffer(2)
print("point_b_buff wkt: {}".format(point_b_buff.wkt))
point_b_buff
```

```python
# point_a buffer boundary
point_a_buff.boundary
```

```python
# point_b_buff centroid
ctx = point_b_buff.centroid
print("point_b_buff wkt: {}".format(ctx.wkt))
```

```python
# Difference point_b_buff - point_a_buff
point_b_buff.difference(point_a_buff)
```

```python
# Difference point_a_buff - point_b_buff
point_a_buff.difference(point_b_buff)
```

```python
# Simmetric difference point_a_buff - point_b_buff
point_a_buff.symmetric_difference(point_b_buff)
```

```python
# Intersection point_a_buff - point_b_buff
point_a_buff.intersection(point_b_buff)
```

```python
# Union point_a_buff - point_b_buff
point_a_buff.union(point_b_buff)
```
