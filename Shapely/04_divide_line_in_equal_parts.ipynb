```python
import numpy as np
from shapely.geometry import LineString, GeometryCollection
import matplotlib.pyplot as plt
```

```python
# Create a line
linea_a = LineString([(-1,0), (0,1)])

# Number of lines
num_lines = 15

segments = []
for i in range(0, num_lines):

    # Get the start point
    start = linea_a.interpolate((i/float(num_lines)), normalized=True)

    # Get the end point
    end = linea_a.interpolate(((i+1)/float(num_lines)), normalized=True)

    # Create the segment
    segment = LineString([start, end])
    segments.append(segment)
```

```python
# Creates a figure and one subplot
fig, ax = plt.subplots()

# Plot each segment
for segnum, segment in enumerate(segments):

   # Get xy coordinates of the segment
   x,y = segment.xy

   # Text position (midpoint)
   midpoint = segment.interpolate(0.5, normalized=True)

   # Plot and annotate the line
   ax.plot(x,y)
   ax.annotate(f"{segnum+1}", (midpoint.x-0.08, midpoint.y),
               color=ax.get_lines()[-1].get_c(), weight='bold')

# Show the plot
ax.set_aspect('equal', adjustable='datalim')
plt.show()
```
