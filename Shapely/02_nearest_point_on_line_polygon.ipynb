**Nearest point on line**

```python
import numpy as np
from shapely.geometry import Point, Polygon, LineString, LinearRing
import matplotlib.pyplot as plt
```

```python
# Start point
point_a = Point(1, 1.4)

# Create a line
line_a = LineString([(1,1), (2,2)])

# Measure the projection point along the line from the starting point
dist_a = line_a.project(point_a)

# Get the closest point
closest_a = line_a.interpolate(dist_a)
print(f"Closest point coordinates: {closest_a.wkt}")
```

```python
# Creates a figure and one subplot
fig, ax = plt.subplots()

# Get xy coordinates of the line
x,y = line_a.xy

# Plot the line
ax.plot(x,y)

# Plot and annotate the start point
ax.scatter(point_a.x, point_a.y, c="r")
ax.annotate(f"Start point:\n{point_a.wkt}", (point_a.x+0.02, point_a.y+0.05))

# Plot and annotate the closest point
ax.scatter(closest_a.x, closest_a.y, c="g")
ax.annotate(f"Nearest point:\n{closest_a.wkt}", (closest_a.x+0.08, closest_a.y-0.05))

# Show the plot
ax.set_aspect('equal', adjustable='datalim')
plt.show()
```

**Nearest point on Polygon**

```python
# Start point
point_b = Point(3, 0)

# Create a Polygon
poly_b = Polygon([(-1,1),(2,1),(2,2),(-1,2)])

# Polygon exterior ring
exterior_ring = LinearRing(poly_b.exterior.coords)

# Measure the projection point along the line from the starting point
dist_b = exterior_ring.project(point_b)

# Get the closest point
closest_b = exterior_ring.interpolate(dist_b)
print(f"Closest point coordinates: {closest_b.wkt}")
```

```python
# Creates a figure and one subplot
fig, ax = plt.subplots()

# Get xy coordinates of the exterior
x_ext, y_ext = exterior_ring.xy

# Plot the polygon
ax.fill(x_ext, y_ext, c="b", alpha=0.2)

# Plot and annotate the start point
ax.scatter(point_b.x, point_b.y, c="r")
ax.annotate(f"Start point:\n{point_b.wkt}", (point_b.x-0.8, point_b.y+0.05))

# Plot and annotate the closest point
ax.scatter(closest_b.x, closest_b.y, c="g")
ax.annotate(f"Nearest point:\n{closest_b.wkt}", (closest_b.x+0.08, closest_b.y+0.05))

# Show the plot
ax.set_aspect('equal', adjustable='datalim')
plt.show()
```
